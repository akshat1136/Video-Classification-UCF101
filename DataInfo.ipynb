```python
num_categories = dict()
for file in (os.listdir('jpegs_256_1/') + os.listdir('jpegs_256_2/') + os.listdir('jpegs_256_3/')):
    cat = file.split('_')[1]
    if cat in num_categories.keys():
        num_categories[cat] += 1
    else:
        num_categories[cat] = 1
```

```python
folder_names = []
for cat in num_categories.keys():
    folder_names = folder_names + [ 'v_' + cat + '_' + str(i) for i in range(num_categories[cat])]
```

```python
for i in range(13320):
    path = 'jpegs_256_'
    if i<4296:
        path = path + '1/'
    elif i<9108:
        path = path + '2/'
    else:
        path = path + '3/'
    os.rename(path + dirs[i], path + folder_names[i])
```

```python
data = []
labels = []
for i in range(101):
    cat = list(num_categories.keys())[i]
    num = num_categories[cat] + 200*i
    labels = labels + list(np.ones(num_categories[cat])*(num//200))
    data.append(np.array(range(200*i, num)))
labels = [int(i) for i in labels]
```

```python
data = np.hstack(data)
len(data), data
```

```python
labels = np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(np.array(range(13320)), test_size = 0.1, shuffle = True)
```

```python
X_train = data[train]
X_test = data[test]
y_train = labels[train]
y_test = labels[test]
```

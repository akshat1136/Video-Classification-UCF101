```python
import sys
import os
import time
from PIL import Image
import numpy as np
import pickle
```

```python
max_frame = 0
min_frame = 20000
for folder in os.listdir('jpegs_256/'):
    frame = len(os.listdir('jpegs_256/'+folder+'/'))
    if frame > max_frame:
        max_frame = frame
        name = folder
    if frame < min_frame:
        min_frame = frame
max_frame, min_frame, name
```

```python
Info = pickle.load(open('Data.p', 'rb'))
Info
```

```python
num_categories, folder_names, Xtrain, ytrain, Xtest, ytest = Info
```

```python
list(num_categories.keys())[30]
```

```python
def folder_name(val, num_categories):
    cat = val//200
    cat = list(num_categories.keys())[cat]
    return cat+'_'+str(val)
```

```python
count = 0
for num in Xtrain:
    fold = folder_name(num, num_categories)
    data = []
    for image in os.listdir('jpegs_256/'+fold+'/'):
        data.append(np.array(Image.open('jpegs_256/'+folder_name(num, num_categories)+'/'+image)))
    count += 1
    print(fold, count)
    data = np.array([data, num//200])
    np.save('NumPy/Train/'+str(fold)+'.npy', data)
```
